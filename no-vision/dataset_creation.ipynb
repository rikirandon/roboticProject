```python
import os
import shutil
from sahi.utils.coco import Coco, CocoCategory, CocoImage, CocoAnnotation
from sahi.utils.file import save_json
import json
from PIL import Image
import ast
```

```python
def rename_images(input_folder):
    for x in range (10):
        image_old_name=f"view={x}.jpeg"
        image_new_name=f"view_{x}.jpeg"

        #destination_folder = f"/home/carlotta/Documents/vision_project/dataset/{y_value}"
        source_path= os.path.join(input_folder, image_old_name)

        destination_path = os.path.join(input_folder, image_new_name)

        os.rename(source_path, destination_path)
```

```python
def create_dataset(input_folder, folder_number, dataset,json_file="view=0.json"):
    os.chdir(input_folder)

    json_filename = json_file  # Sostituisci con il nome del tuo file JSON
    obj_attr = 'obj-0'
    y_attr = 'y'
    bbox_attr = "3d_bbox"

    with open(json_filename, 'r') as file:
        data = json.load(file)
        #print(data)

    if obj_attr in data and y_attr in data[obj_attr]:
        y_value = data[obj_attr][y_attr]
        print(f"The value of '{obj_attr}.{y_attr}' is: {y_value}")
    else:
        print(f"Nested attribute '{obj_attr}.{y_attr}' not found in the JSON file.")

    if obj_attr in data and bbox_attr in data[obj_attr]:
        bbox_coords = data[obj_attr][bbox_attr]
        print(f"The value of '{obj_attr}.{y_attr}' is: {bbox_coords}")
    else:
        print(f"Nested attribute '{obj_attr}.{bbox_coords}' not found in the JSON file.")


    for x in range(10):
        image_old_name=f"view_{x}.jpeg"
        image_new_name=f"{y_value}-{x}-{folder_number}.jpeg"

        new_folder_path = f"/home/carlotta/Documents/vision_project/{dataset}/{y_value}/"
        new_folder_name = f"{x}-{folder_number}"

        destination_folder = os.path.join(new_folder_path, new_folder_name)

        os.makedirs(destination_folder, exist_ok=True)

        source_path= os.path.join(input_folder, image_old_name)

        destination_path = os.path.join(destination_folder, image_new_name)

        file_path=os.path.join(destination_folder, f"coordinates.{x}-{folder_number}.txt")

        try:
        # Rename the file
            os.rename(source_path, destination_path)

            with open(file_path, "w") as file:
                file.write(f"{bbox_coords}\n")
        # Move the renamed image to the destination folder
            shutil.move(destination_path, destination_folder)

            print(f"Image renamed and moved successfully to {destination_path}")
        except FileNotFoundError:
            print(f"Error: The file {source_path} does not exist.")
        except Exception as e:
            print(f"Error: {e}")
```

```python
for x in range (1, 51):
    rename_images(f"/home/carlotta/Documents/vision_project/assign1/scene{x}")
```

```python
for x in range(1, 51):
    create_dataset(f"/home/carlotta/Documents/vision_project/assign1/scene{x}", x, dataset="dataset_v4")

```
